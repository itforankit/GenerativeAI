```python
import tensorflow as tf
from tensorflow import keras
```

```python
mnist = keras.datasets.mnist
```

```python
(X_train, Y_train), (X_test, Y_test) =mnist.load_data()
```

```python
X_train[2]
```

```python
RESHAPE=784
X_train = X_train.reshape(60000,RESHAPE)
X_test = X_test.reshape(10000,RESHAPE)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.astype('float')
X_test.astype('float')
```

```python
X_train=X_train/255
X_test =X_test/255
```

```python
X_train[2]
```

```python
Y_train = tf.keras.utils.to_categorical(Y_train,10)
Y_test = tf.keras.utils.to_categorical(Y_test,10)
```

```python
Y_train[0]
```

design the model
define the architecure of the model

Sequential Model -> 1 hidden layer
stack of layers

```python
model = tf.keras.Sequential()
```

```python
# 1 hidden layer 64 neuorns
noofhidden=64
model.add(tf.keras.layers.Dense(noofhidden, activation='relu'))

# 2nd hidden layer will have 30 neuorns
model.add(tf.keras.layers.Dense(30, activation='relu'))
```

```python
#addd output layer with 10 output units/neuorns
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
#compiling the model
import keras.losses
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#train the model
#60000/64=938
model.fit(X_train, Y_train, batch_size=64, epochs=50)
```

```python
model.evaluate(X_test,Y_test)
```
