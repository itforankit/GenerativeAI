# Step-3: Prepare the Sentiment Analysis Pipeline and Define Texts and Analyze Sentiment

```python
from transformers import pipeline

# Initialize a pipeline for sentiment analysis
sentiment_analysis = pipeline("sentiment-analysis")

# Example texts
texts = ["I love this product!", "This is the worst experience ever.", "It's okay, not great, but not bad either."]

# Analyze sentiment for each text
for text in texts:
    result = sentiment_analysis(text)
    print(f"Text: {text}\nSentiment: {result[0]['label']}, Confidence: {result[0]['score']:.2f}\n")
```
